## Triadex Muse

The settings below are the worked example from the original programming manual. The result is still an ordinary Paeonia `Bar`.

```python
muse = TriadexMuse(
    interval=("C1", "B4", "B2", "B7"),
    theme=("OFF", "B1", "B3", "B6"),
    root=60,
    rest=True,
)

phrase = muse.to_bar(beats=32, unit=Fraction(1, 8))
phrase
```

```python
# Requires FluidSynth.
phrase.play()
```
